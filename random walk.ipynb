```python
import random
import matplotlib.pyplot as plt
import numpy
```

```python
def random_walk(X_min, X_max, y_min, y_max, n, it): 
    X_range = X_max - X_min
    y_range = y_max - y_min
    
    # random particle spawn
    x = np.random.randint(X_min, X_max, size=(1, n))
    y = np.random.randint(y_min, y_max, size=(1, n))
    
    # new pos 
    new_x = np.zeros((1, n))
    new_y = np.zeros((1, n))
    for i in range(it): 
        for particle in range(n):
            # probability
            rand = random.randint(1, 4)
            if rand == 1: 
                new_x[0][particle] = x[i][particle] + 1
            elif rand == 2:
                new_x[0][particle] = x[i][particle] - 1
            elif rand == 3:
                new_y[0][particle] = y[i][particle] + 1
            else : 
                new_y[0][particle] = y[i][particle] - 1
                
            # PBC 
            if new_x[0][particle] > X_max : 
                new_x[0][particle] = new_x[0][particle] - X_range
                
            if new_x[0][particle] < X_min : 
                new_x[0][particle] = new_x[0][particle] + X_range
                
            if new_x[0][particle] > y_max : 
                new_y[0][particle] = new_x[0][particle] - y_range
                
            if new_x[0][particle] < y_min : 
                new_y[0][particle] = new_x[0][particle] + y_range

        x = np.concatenate((x, new_x))
        y = np.concatenate((y, new_y))
    
    return x, y

x, y = random_walk(-25, 25, -25, 25 ,10 , 100)


# plot 
fig = plt.figure(figsize=(15,15))
plt.scatter(x[0],y[0], color="r", label="start point")
plt.scatter(x[-1],y[-1], color="blue", label="end point")
plt.plot(x,y, label ='path')
plt.legend()
```
